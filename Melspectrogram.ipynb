```python
def wav_mel(file):
    array, rate = librosa.load(file)
    array = array[0:66150]

    print(array)
    print(rate)
    X = librosa.stft(array)
    Xdb = librosa.amplitude_to_db(abs(X))
    librosa.display.specshow(Xdb, sr=rate, x_axis='time', y_axis='hz')
```

```python
def wav_mel(file):
    fig, ax = plt.subplots()
    s, rate = librosa.load(file)
    s = s[0:66150]
    
    mel = librosa.feature.melspectrogram(y=s, sr=rate)
    S_dB = librosa.power_to_db(mel, ref=np.max)
    img = librosa.display.specshow(S_dB, x_axis='time',
                         y_axis='mel', sr=rate,
                         fmax=8000, ax=ax)
```
